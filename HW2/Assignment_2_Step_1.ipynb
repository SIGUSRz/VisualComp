```matlab
% Part 1

% Part A

% Step 1
img = zeros(512,512);

% Step 2

% x is row, y is col
for row=1:size(img,1)
    for col=1:size(img,2)
        img(row, col) = sin(0.2*row) + sin(0.3*row) + cos(0.4*row) + sin(0.15*(row^2 + col^2)^0.5) + sin(0.15*(row^2 + col^2)^0.5);
    end
end

imshow(img)
```

```matlab
% Step 3

dft = fft2(img);
imshow(real(dft))
```

```matlab
m = abs(dft);                               % Magnitude
imagesc(abs(fftshift(m)))
```

```matlab
dft(m<1e-6) = 0;
p = unwrap(angle(dft));                     % Phase
imagesc(abs(fftshift(p)))
```

```matlab
% Step 4

dft = fft2(img);
% imagesc(abs(fftshift(dft)))
idft = ifft2((2*abs(dft)));
% imagesc(idft)
imshow(idft)

% It's a zoomed out version of the first one. Now it has the same pattern but twice
```
